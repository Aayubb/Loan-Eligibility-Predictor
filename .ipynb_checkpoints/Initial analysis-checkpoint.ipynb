##### The column `Gender`, `Married`, `Dependents`, `Self_Employed`, `LoanAmount`, `Loan_Amount_Term` and `Credit_History` has missing values.

To fill these missing values, we need to apply two different methods.

- Fill the numerical columns with the median (the middle value) - `LoanAmount`, `Loan_Amount_Term`.
- Fill the categorical columns with mode (most frequent value) - `Gender`, `Married`, `Dependents`,  `Self_Employed`

```python
# Filling missing numerical values with median

#Loan amount
loan_amt_mid =train_ds['LoanAmount'].median()
train_ds['LoanAmount'] = train_ds["LoanAmount"].fillna(loan_amt_mid)


# Loan_Amount_Term
loan_amt_term_mid =train_ds['Loan_Amount_Term'].median()
train_ds['Loan_Amount_Term'] = train_ds["Loan_Amount_Term"].fillna(loan_amt_term_mid)

# Credit History
credit_history_mid = train_ds['Credit_History'].median()
train_ds['Credit_History'] = train_ds['Credit_History'].fillna(credit_history_mid)
```

```python
# Filling categorical columns with mode

# Gender
gender_md = train_ds['Gender'].mode()[0]
train_ds['Gender'] = train_ds['Gender'].fillna(gender_md)

# Married
married_md = train_ds['Married'].mode()[0]
train_ds['Married'] = train_ds['Married'].fillna(married_md)

#Dependents
dependents_md = train_ds['Dependents'].mode()[0]
train_ds['Dependents'] = train_ds['Dependents'].fillna(dependents_md)

# Self_employed
self_employed_md = train_ds['Self_Employed'].mode()[0]
train_ds['Self_Employed'] = train_ds['Self_Employed'].fillna(self_employed_md)
```
